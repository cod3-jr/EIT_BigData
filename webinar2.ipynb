```python
import matplotlib.pyplot as plt
import math
def mean(dataset): 
    total=0
    for element in dataset:
        total=total+element
    return total/len(dataset)

def variance(dataset):
    total=0
    setmean=mean(dataset)
    for element in dataset:
        total=total+(math.pow(element-setmean,2))
    return total/len(dataset)

mdata = [134045,102343,108279,127634,117738,125018,103488]
mdata2 = [34045,102343,279,127634,11738,155018,393488]
mmean = mean(mdata)
mmean2 = mean(mdata2)
print("Variance1 = " + str(variance(mdata)))
print("Variance2 = " + str(variance(mdata2)))
print(mmean)
print(mmean2)
plt.plot(mdata,label='Profits')
plt.plot(mdata2,label='Profits2')
plt.plot([mmean]*7,label='Mean')
plt.plot([mmean2]*7,label='Mean2')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

mu = 100 # mean
sigma = 100 # standard deviation
s = np.random.normal(mu,sigma,10000)
s2 = np.random.normal(mu,10,10000)
#plt.hist(s)
#you might also force the number of bins to be bigger
plt.hist(s,100)
plt.hist(s2,100)

#you can also normalize
plt.hist(s,100,density=True)
plt.hist(s2,100,density=True,cumulative=True)
```
